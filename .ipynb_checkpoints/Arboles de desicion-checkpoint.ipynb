```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split

%matplotlib inline
sns.set()
```

```python
train_df = pd.read_csv('titanic-train.csv')
test_df = pd.read_csv('titanic-test.csv')
train_df.head()
```

```python
train_df.info()
```

```python
train_df.Sex.value_counts().plot(kind = 'bar', color = ['b','r'])
plt.title('Distribucion de sexos')
plt.show()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

encoder_sex = label_encoder.fit_transform(train_df['Sex'])
train_df['Sex'] = encoder_sex
train_df.head()
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
train_df['Embarked'] = train_df['Embarked'].fillna('S')
```

```python
train_predictor = train_df.drop(['PassengerId','Survived','Name','Ticket','Cabin'], axis = 1)
categorical_cols = [cname  for cname in train_predictor.columns
                   if train_predictor[cname].nunique() < 10  and
                   train_predictor[cname].dtype == 'object']
```

```python
numerical_cols = [cname for cname in train_predictor.columns if
                 train_predictor[cname].dtype in ['int64','float64']]
```

```python
my_cols = categorical_cols + numerical_cols

my_cols
```

```python
train_predictor = train_predictor[my_cols]
train_predictor
```

```python
dummy_enconded_train_predictor = pd.get_dummies(train_predictor)
dummy_enconded_train_predictor
```

```python
train_df['Pclass'].value_counts()
```

```python
y_target = train_df['Survived'].values
x_features_one = dummy_enconded_train_predictor.values
```

```python
x_train, x_validation, y_train, y_validation = train_test_split(x_features_one, y_target, test_size = 0.25, random_state = 1)
```

```python
tree_one = tree.DecisionTreeClassifier()
tree_one = tree_one.fit(x_train, y_train)
```

```python
tree_one_accuracy = round(tree_one.score(x_validation, y_validation), 4)
print(tree_one_accuracy)
```
